```
def ohe(data, col_name):
    ohe = OneHotEncoder();
    array = ohe.fit_transform(data[[col_name]]).toarray()
    unique_col_names = sorted(list(pd.unique(data[col_name])))
    unique_col_names_final = np.array([])
    for name in unique_col_names:
        unique_col_names_final = np.append(unique_col_names_final, f'{col_name}_{name}')
    temp_df = pd.DataFrame(array, columns=unique_col_names_final)
    temp_df.index = data.index
    new_df = data.join(temp_df)
    return new_df, unique_col_names_final.tolist()
```

```
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import regex as re

def ohe_list(data, column_name):
    """
    One-hot encodes the specified column in the given DataFrame and retains all original columns in addition to the 'Name' column and the one-hot encoded columns.

    :param data: A pandas DataFrame containing the data to be encoded.
    :param column_name: The name of the column to be one-hot encoded. This column should contain lists.
    :return: A DataFrame with all original columns, the 'Name' column, and the one-hot encoded columns.
    """
    df = pd.DataFrame(data).copy()

    # Define the regular expression pattern for splitting the strings
    pattern = r'[,&+]+|and'

    # Remove "[" and "]" from the column before splitting
    df[column_name] = df[column_name].apply(lambda x: re.sub(r'[\[\]]', '', str(x)))

    # Split the column based on the defined pattern and ensure it is a list
    df[column_name] = df[column_name].apply(lambda x: [re.sub(r'\s+', '', item.lower()) for item in re.split(pattern, str(x))])

    # Instantiate the OneHotEncoder
    encoder = OneHotEncoder(sparse=False)

    # Flatten the list of values into a Series object and create a new DataFrame for them
    values_series = df[column_name].apply(pd.Series).stack().reset_index(level=1, drop=True)
    values_df = df.drop(columns=[column_name]).join(pd.DataFrame(values_series, columns=[column_name])).reset_index()

    # Apply one-hot encoding
    values_encoded = encoder.fit_transform(values_df[[column_name]])
    values_encoded_df = pd.DataFrame(values_encoded, columns=encoder.get_feature_names_out([column_name]))

    # Combine the one-hot encoded data with the original index
    values_encoded_df['index'] = values_df['index']
    values_encoded_df = values_encoded_df.groupby('index').sum()

    # Concatenate the original DataFrame with the one-hot encoded DataFrame
    df_final = pd.concat([df, values_encoded_df], axis=1)

    return df_final
```

```
import pandas as pd
import networkx as nx
import torch
from torch_geometric.data import Data
from itertools import combinations
# Update data with negative edges
from torch_geometric.utils import negative_sampling, train_test_split_edges
from torch_geometric.nn import Sequential, GCNConv, GATConv, BatchNorm
import torch.nn.functional as F

# Assuming 'df_ohe' is your one-hot encoded pandas DataFrame and it has columns 'Name', 'friend_1', 'friend_2', ..., 'friend_5'

# Create a hashmap from the name to the index number
name_to_index_hashmap = {row['Name']: index for index, row in df_ohe_only.iterrows()}
print(name_to_index_hashmap["Arthur Baruzzi"])

# Create a weighted, undirected graph
G = nx.Graph()

# Add nodes to the graph using index numbers
for index in df.index:
    name = df.loc[index, 'Name']# Get the name from the DataFrame
    G.add_node(index, name=name)

# Add edges to the graph based on friend columns 'friend_1' to 'friend_5'
for index, row in df.iterrows():
    for i in range(1, 6):  # Assuming there are 5 friend columns
        friend_name = row.get(f'friend_{i}')
        if pd.notna(friend_name) and friend_name in name_to_index_hashmap:
            friend_index = name_to_index_hashmap[friend_name]
            weight = 1.0 - (i - 1) * 0.2
            G.add_edge(index, friend_index, weight=weight)  # placeholder algorithm for weight depends on friend number

# Generate negative edges for every possible relationship without a positive edge with the other person already
all_node_pairs = list(combinations(range(len(G.nodes())), 2))
negative_edges = [(u, v) for (u, v) in all_node_pairs if not G.has_edge(u, v)]
print([(u, v) for (u, v) in all_node_pairs if (u,v) not in negative_edges])

# Convert negative edges to torch.tensor
neg_edge_index = torch.tensor(negative_edges).t().contiguous()
print(neg_edge_index.shape)

# Now, neg_edge_index contains all negative edges without a corresponding positive edge

# Assume each node has features, use real feature data here
number_of_features = 54  # Replace with the actual number of features
# Extract features from the one-hot encoded DataFrame
features_data = df_ohe_only.drop(columns=['Name'])
features_array = features_data.to_numpy()  # Convert to numpy array
features = torch.tensor(features_array, dtype=torch.float)
print(features.shape) # Convert to PyTorch tensor

# Create edge_index from the graph edges
edges = list(G.edges())
edge_index = torch.tensor([[edge[0], edge[1]] for edge in edges], dtype=torch.long).t().contiguous()
print(edge_index.shape)


# Create a list of edge weights
edge_weights = torch.tensor([G[u][v]['weight'] for u, v in G.edges], dtype=torch.float)

# Construct PyTorch Geometric data object
data = Data(x=features, edge_index=edge_index, edge_attr=edge_weights)
```

```

from torch_geometric.utils import negative_sampling, train_test_split_edges

# Use PyTorch Geometric's utility functions to split the edges
data2 = train_test_split_edges(data, val_ratio = 0,
    test_ratio= 0.2,)
# Calculate split sizes
total_cols = neg_edge_index.size(1)
split_col = int(total_cols * 0.8)

# Generate a random permutation of indices
indices = torch.randperm(total_cols)

# Split indices for two subsets
train_indices = indices[:split_col]
test_indices = indices[split_col:]

# Use the indices to split the tensor
train_tensor = neg_edge_index[:, train_indices]
test_tensor = neg_edge_index[:, test_indices]

data2.train_neg_edge_index= train_tensor
data2.test_neg_edge_index = test_tensor

print(f"Positive training edges: {data.train_pos_edge_index.size(1)}")
print(f"Negative training edges: {data.train_neg_edge_index.size(1)}")

# class GCNLinkPrediction(torch.nn.Module):
#     def __init__(self, num_features, hidden_dim):
#         super(GCNLinkPrediction, self).__init__()
#         self.conv1 = GCNConv(num_features, hidden_dim)
#         self.conv2 = GCNConv(hidden_dim, hidden_dim)

#     def forward(self, data):
#         x, edge_index = data.x, data.train_pos_edge_index
#         x = self.conv1(x, edge_index)
#         x = F.relu(x)
#         x = self.conv2(x, edge_index)
#         return x

class GCNLinkPrediction(torch.nn.Module):
    def __init__(self, num_features, hidden_dim):
        super(GCNLinkPrediction, self).__init__()
        self.conv1 = GCNConv(num_features, hidden_dim)
        self.bn1 = BatchNorm(hidden_dim)
        self.conv2 = GCNConv(hidden_dim, hidden_dim)
        self.bn2 = BatchNorm(hidden_dim)

    def forward(self, data):
        x, edge_index = data.x, data.train_pos_edge_index
        x = self.conv1(x, edge_index)
        x = self.bn1(x)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        x = self.bn2(x)
        return x

# class GCNLinkPrediction(torch.nn.Module):
#     def __init__(self, num_features, hidden_dim, dropout=0.5):
#         super(GCNLinkPrediction, self).__init__()
#         self.conv1 = GCNConv(num_features, hidden_dim)
#         self.conv2 = GCNConv(hidden_dim, hidden_dim)
#         self.dropout = dropout

#     def forward(self, data):
#         x, edge_index = data.x, data.train_pos_edge_index
#         x = F.dropout(x, p=self.dropout, training=self.training)
#         x = F.relu(self.conv1(x, edge_index))
#         x = F.dropout(x, p=self.dropout, training=self.training)
#         x = self.conv2(x, edge_index)
#         return x

# # Instantiate the model
model = GCNLinkPrediction(num_features=54, hidden_dim=64)

# class GATLinkPrediction(torch.nn.Module):
#     def __init__(self, num_features, hidden_dim, heads=8, dropout=0.6):
#         super(GATLinkPrediction, self).__init__()
#         # First GAT layer
#         self.conv1 = GATConv(num_features, hidden_dim, heads=heads, dropout=dropout)
#         # Output layer, reduce number of heads to 1 for predictions
#         self.conv2 = GATConv(hidden_dim * heads, hidden_dim, heads=1, concat=False, dropout=dropout)
#         self.dropout = dropout

#     def forward(self, data):
#         x, edge_index = data.x, data.train_pos_edge_index
#         x = F.dropout(x, p=self.dropout, training=self.training)  # Apply dropout to input features
#         x = F.relu(self.conv1(x, edge_index))  # First GAT layer with ReLU activation
#         x = F.dropout(x, p=self.dropout, training=self.training)  # Dropout before the second GAT layer
#         x = self.conv2(x, edge_index)  # Second GAT layer
#         return x
# model = GATLinkPrediction(num_features=54, hidden_dim=16, heads=6, dropout=0.4)

def predict_links(node_embeddings, edge_index):
    # Calculate the dot product between each pair of nodes in edge_index
    edge_scores = torch.sum(node_embeddings[edge_index[0]] * node_embeddings[edge_index[1]], dim=1)
    return torch.sigmoid(edge_scores)  # Use sigmoid to get probabilities

# Example usage
model.train()
optimizer = torch.optim.Adam(model.parameters(), lr=0.005, weight_decay=5e-4) 
criterion = torch.nn.BCEWithLogitsLoss()

print("Feature shape:", data2.x.shape)  # Expect [num_nodes, num_features]
print("Positive edge index shape:", data2.train_pos_edge_index.shape)  # Expect [2, num_positive_edges]
print("Negative edge index shape:", data2.train_neg_edge_index.shape)  # Expect [2, num_negative_edges]

for epoch in range(500):
    optimizer.zero_grad()
    node_embeddings = model(data2)
    # Compute the scores for the positive edges
    # torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
    pos_edge_scores = predict_links(node_embeddings, data2.train_pos_edge_index)
    # Compute the scores for the negative edges
    neg_edge_scores = predict_links(node_embeddings, data2.train_neg_edge_index)
    # Labels: ones for positive edges, zeros for negative edges
    labels = torch.cat([torch.ones(pos_edge_scores.size(0)), torch.zeros(neg_edge_scores.size(0))], dim=0)
    # Scores: concatenate positive and negative scores
    scores = torch.cat([pos_edge_scores, neg_edge_scores], dim=0)
    loss = criterion(scores, labels)
    torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")
```

```
from sklearn.metrics import roc_auc_score

model.eval()
with torch.no_grad():
    node_embeddings = model(data2)
    test_pos_edge_scores = predict_links(node_embeddings, data2.train_pos_edge_index)
    test_neg_edge_scores = predict_links(node_embeddings, data2.train_neg_edge_index)

    # Concatenate positive and negative scores, and create corresponding labels
    test_scores = torch.cat([test_pos_edge_scores, test_neg_edge_scores]).cpu().numpy()
    test_labels = torch.cat([torch.ones(test_pos_edge_scores.size(0)), torch.zeros(test_neg_edge_scores.size(0))]).cpu().numpy()

    # Calculate ROC AUC
    auc_score = roc_auc_score(test_labels, test_scores)
    print("Test ROC AUC Score:", auc_score)
```

```
def get_connected_nodes(data, target_node):
    # Initialize a set to avoid duplicates
    connected_nodes = set()

    # Check if training positive edges are available and find connections
    if hasattr(data, 'train_pos_edge_index'):
        # Extract nodes connected to target_node in training data
        mask = (data.train_pos_edge_index[0] == target_node) | (data.train_pos_edge_index[1] == target_node)
        connected_train_nodes = data.train_pos_edge_index[:, mask]
        # Add to set (both directions could be relevant)
        connected_nodes.update(connected_train_nodes[0].tolist())
        connected_nodes.update(connected_train_nodes[1].tolist())

    # Check if testing positive edges are available and find connections
    if hasattr(data, 'test_pos_edge_index'):
        # Extract nodes connected to target_node in testing data
        mask = (data.test_pos_edge_index[0] == target_node) | (data.test_pos_edge_index[1] == target_node)
        connected_test_nodes = data.test_pos_edge_index[:, mask]
        # Add to set (both directions could be relevant)
        connected_nodes.update(connected_test_nodes[0].tolist())
        connected_nodes.update(connected_test_nodes[1].tolist())

    return connected_nodes
def recommend_new_links(model, data, target_node):
    model.eval()
    with torch.no_grad():
        node_embeddings = model(data)  # Generate node embeddings using the trained model

    target_embedding = node_embeddings[target_node]  # Get the embedding of the target node
    similarity_scores = (node_embeddings @ target_embedding).cpu().numpy()

    # Get indices of existing connections to filter them out
    existing_connections = get_connected_nodes(data, target_node)
    recommendations = []

    for i, score in enumerate(similarity_scores):
        # Skip if it's the target node itself or if an edge already exists
        if i == target_node or i in existing_connections:
            continue
        recommendations.append((i, score))

    # Sort based on scores, descending, and return the top N recommendations
    recommendations.sort(key=lambda x: x[1], reverse=True)
    return recommendations[:3]  # Adjust N as needed

# Example usage for a target node
# target_node = 2  # Example target node
# recommendations = recommend_new_links(model, data2, target_node)
# print("Recommended new links for node", target_node, ":", recommendations)
```
